```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
data = []
for i in range(1,51):
    url = f'http://books.toscrape.com/catalogue/page-{i}.html'
    page = requests.get(url)
    soup = BeautifulSoup(page.text,'html.parser')
    all_books = soup.find_all('li',class_= 'col-xs-6 col-sm-4 col-md-3 col-lg-3')
    for books in all_books:
        item = {}
        item['Title'] = books.find('img').get('alt','')
        item['Link'] = books.find('a').get('href','')
        item['price'] = books.find('p', class_= 'price_color').text.strip()[2:]
        item['stock'] = books.find('p', class_= 'instock availabilty')
        data.append(item)
        df = pd.DataFrame(data)
        df.to_csv(r'C:\Users\HP\Documents\New folder\scraped1.csv',index = False)
        print(f'scraped {i} pages successfully ')
        
    
```
